```python
import requests


api_url = "https://api.mangaupdates.com/v1"

def login():
    login_endpoint = f"{api_url}/account/login"

    login_data = {
        "username": "omlnaut",
        "password": "redacted"
    }

    response = requests.put(login_endpoint, data=login_data).json()
    return response['context']['session_token']
```

```python
token = login()
token
```

```python
search_endpoint = f"{api_url}/series/search"
payload = {
    "search": "Omniscient Reader",}

search_response = requests.post(search_endpoint, json=payload, headers={"Authorization": f"Bearer {token}"}).json()
```

```python
series_id = "50369844984"
series_endoint = f"{api_url}/series/{series_id}"

series_response = requests.get(series_endoint, headers={"Authorization": f"Bearer {token}"}).json()
series_response
```

```python
releases_endpoint = f"{api_url}/releases/search"

payload = {"search": "50369844984","search_type": "series",  "asc": "desc", "group_id": 57949066600, "include_metadata": False}

releases_response = requests.post(releases_endpoint, json=payload, headers={"Authorization": f"Bearer {token}"}).json()
releases_response
```
